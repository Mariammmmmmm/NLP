```python
import re
import string
import nltk
from nltk.corpus import wordnet 
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

```python
import pandas as pd
data = pd.read_csv('financial_review.csv')
data.tail(10)
```

```python
text = data['review'].to_string()
type(text)
```

```python
from nltk.tokenize import word_tokenize
# convert to lower case, removing stop words, numbers / non-alphabetic characters, lemmatization.

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')

stop_words = set(stopwords.words('russian'))
lemmatizer = WordNetLemmatizer()
def clear(text):
  text = text.lower()
  text = list(filter(str.isalpha, word_tokenize(text.lower())))
  text = list(lemmatizer.lemmatize(word) for word in text)
  text = list(word for word in text if word not in stop_words)
  new_text = ' '.join(text)
  return new_text

text = clear(text)
text[:500]
```

```python
from nltk.tokenize import word_tokenize
text_word = word_tokenize(text)
```

```python
#10 популярных слов
from nltk.probability import FreqDist
fdist = FreqDist()

for word in text_word:
  fdist[word.lower()]+=1

fdist_top10 = fdist.most_common(10)
fdist_top10
```

Разделим отзывы на положительные и отрицательные

```python
# Создание столбца по определению положительных/отрицательных отзывов
data = data.dropna(subset=['score'])
filter = (data['score'] != '1') & (data['score'] != '2') & (data['score'] != '4') & (data['score'] != '5')
df = data.drop(index=data[filter].index)
df['scores'] = ['positive' if  (( i== '4')|(i == '5')) else 'negative' for i in df.score]
df[['review','score', 'scores']]
```

```python
#Поделим на 2 датафрейма с положительными и отрицательными отзывами
filter = (df['scores'] == 'positive')
dfneg = df.drop(index=df[filter].index)
filter = (df['scores'] == 'negative')
dfpoz = df.drop(index=df[filter].index)
dfpoz[['review','score', 'scores']]
```

Топ слов в позитивных отзывах

```python
import re
positive_review = dfpoz['review'].to_string()

positive_review = positive_review.lower()
positive_review = word_tokenize(positive_review)
positive_review = [word for word in positive_review if word.isalpha()]
positive_review = list(lemmatizer.lemmatize(word) for word in positive_review)
positive_review = list(word for word in positive_review if word not in stop_words)
positive_review = list(positive_review[i] for i in range(len(positive_review)-1) if positive_review[i+1]=='банк')
positive_review = str(positive_review)
positive_review = re.findall(r'\w+й', positive_review)
positive_review = ' '.join(positive_review)
positive_review[:500]


positive_review = word_tokenize(positive_review)
#10 популярных слов позитивного интервью
posfdist = FreqDist()

for word in positive_review:
  posfdist[word.lower()]+=1

posfdist_top10 = posfdist.most_common(10)
posfdist_top10
```

Топ слов в негативных отзывах

```python
negative_review = dfneg['review'].to_string()

negative_review = negative_review.lower()
negative_review = word_tokenize(negative_review)
negative_review = [word for word in negative_review if word.isalpha()]
negative_review = list(lemmatizer.lemmatize(word) for word in negative_review)
negative_review = list(word for word in negative_review if word not in stop_words)
negative_review = list(negative_review[i] for i in range(len(negative_review)-1) if negative_review[i+1]=='банк')
negative_review = str(negative_review)
negative_review = re.findall(r'\w+й', negative_review)
negative_review = ' '.join(negative_review)
negative_review[:500]


negative_review = word_tokenize(negative_review)
#10 популярных слов позитивного интервью
negfdist = FreqDist()

for word in negative_review:
  negfdist[word.lower()]+=1

negfdist_top10 = negfdist.most_common(10)
negfdist_top10
```
